```python
import pandas as pd
import numpy as np 
```

```python
btc_perp = pd.read_csv("BinanceU_BTCUSDC_perp.csv")
btc_funding = pd.read_csv("Funding_BinanceU_BTCUSDC_perp.csv")

btc_perp['timestamp'] = btc_perp['open_time'].apply(lambda t: pd.to_datetime(int(t/1000), unit='s')).astype(str)
btc_funding['timestamp'] = btc_funding['calc_time'].apply(lambda t: pd.to_datetime(int(t/1000), unit='s')).astype(str)
```

```python
btc_perp.to_csv("BinanceU_BTCUSDC_perp.csv", index=False)
btc_funding.to_csv("Funding_BinanceU_BTCUSDC_perp.csv", index=False)
```
